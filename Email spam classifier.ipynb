```python
spam_keywords = ["win", "free", "offer", "money", "prize"]
emails = [
    "You win a free prize",
    "Meeting tomorrow at 10",
    "Limited time offer just for you",
    "Project discussion update"
]
for email in emails:
    spam_score = 0
    for word in spam_keywords:
        if word in email.lower():
            spam_score += 1
    if spam_score > 0:
        print(email, "-> SPAM")
    else:
        print(email, "-> NOT SPAM")
```

```python
email = input("Enter email message: ")
spam_keywords = ["win", "free", "offer", "money", "prize"]
spam_score = 0
for word in spam_keywords:
    if word in email.lower():
        spam_score += 1
if spam_score > 0:
    print("This email is SPAM")
else:
    print("This email is NOT SPAM")
```
